```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

# 球の半径
radius = 1

# 描画する面（網目）の数
u_res = 50
v_res = 50

# 角度の範囲
u_min = 0
u_max = 2 * np.pi
v_min = 0
v_max = np.pi

# 面上の座標を計算
u = np.linspace(u_min, u_max, u_res)
v = np.linspace(v_min, v_max, v_res)
u_mesh, v_mesh = np.meshgrid(u, v)

# デカルト座標に変換
x_mesh = radius * np.cos(u_mesh) * np.sin(v_mesh)
y_mesh = radius * np.sin(u_mesh) * np.sin(v_mesh)
z_mesh = radius * np.cos(v_mesh)

# 図の準備
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 球の描画 (面)
ax.plot_surface(x_mesh, y_mesh, z_mesh, color='skyblue', alpha=0.7)

# 球の描画 (ワイヤーフレーム)
# ax.plot_wireframe(x_mesh, y_mesh, z_mesh, color='black', linewidth=0.5)

# 軸の表示
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# 軸の範囲を設定（任意）
# ax.set_xlim([-radius*1.5, radius*1.5])
# ax.set_ylim([-radius*1.5, radius*1.5])
# ax.set_zlim([-radius*1.5, radius*1.5])

# 図の表示
plt.show()
```

コード解説:
1. ライブラリのインポート:
matplotlib.pyplot、numpy、mpl_toolkits.mplot3dをインポートします。
2. 球の半径の設定:
radius変数に球の半径を設定します。
3. 描画面の定義:
u_resとv_resで、球の表面を分割する面の数（分割数が多いほど滑らかな球になります）。
4. 角度の範囲:
u_minからu_maxまでをlongitude（経度）、v_minからv_maxまでをlatitude（緯度）として設定します。
5. 座標の計算:
`np.linspace()`で角度を分割し、`np.meshgrid()`で格子状の座標を生成します。
6. デカルト座標への変換:
x_mesh, y_mesh, z_meshで、球座標（経度、緯度）をデカルト座標（x, y, z）に変換します。
7. 図の準備:
`plt.figure()`で図のフレームを作成し、`fig.add_subplot()`で3次元プロット用のAxesオブジェクトを作成します。
8. 球の描画:
`ax.plot_surface()`で球の面を描画します。`ax.plot_wireframe()`でワイヤーフレームを描画することも可能です。
9. 軸の設定:
軸ラベルを付与し、必要に応じて軸の範囲を設定します。
10. 図の表示:
`plt.show()`で図を表示します。
注意点:
`plot_surface()`は、指定した色の塗り潰された面を描画します。
`plot_wireframe()`は、球の輪郭を描画します。
軸の範囲を設定することで、球の表示を調整できます。
描画面の数（u_resとv_res）を大きくすることで、より滑らかな球を描画できます。
